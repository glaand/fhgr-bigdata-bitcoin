# Visualisierungen BTC Preise

```python
# imports
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

## Data

```python
daily_df = pd.read_csv("prices_daily.csv")
hourly_df = pd.read_csv("prices_hourly.csv")
```

“Volume From” and “To” are the volumes of the respective currency pair. <br>
For example, for the BTC-USD pair “Volume From” is the number of Bitcoins traded for US dollars <br>
while “Volume To” is the number of dollars traded (for the period) for Bitcoins.

## Plots

### Daily

```python
# Set the figure size and create the subplots
fig = plt.figure(figsize=(25,15))
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax6 = fig.add_subplot(326)

# Convert the time to Matplotlib's date format
date_time = mdates.date2num(daily_df.time)

# Plot the data on the subplots
ax1.plot(date_time, daily_df.high)
ax1.set_title("Daily BTC High Price")
ax1.set_xlabel("Time")
ax1.set_ylabel("Price [USD]")
ax1.grid()

ax2.plot(date_time, daily_df.low)
ax2.set_title("Daily BTC Low Price")
ax2.set_xlabel("Time")
ax2.set_ylabel("Price [USD]")
ax2.grid()

ax3.plot(date_time, daily_df.close)
ax3.set_title("Daily BTC Close Price")
ax3.set_xlabel("Time")
ax3.set_ylabel("Price [USD]")
ax3.grid()

ax4.plot(date_time, daily_df.open)
ax4.set_title("Daily BTC Open Price")
ax4.set_xlabel("Time")
ax4.set_ylabel("Price [USD]")
ax4.grid()

ax5.plot(date_time, daily_df.volumefrom)
ax5.set_title("Daily number of Bitcoins traded for USD")
ax5.set_xlabel("Time")
ax5.set_ylabel("#BTC")
ax5.grid()

ax6.plot(date_time, daily_df.volumeto)
ax6.set_title("Daily number of USD traded for Bitcoins")
ax6.set_xlabel("Time")
ax6.set_ylabel("#USD")
ax6.grid()

# Format the x-axis date labels
for ax in fig.axes:
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.suptitle("BTC Daily")

# Rotate the x-axis tick labels
plt.gcf().autofmt_xdate()

# Adjust the subplots to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

### Hourly
